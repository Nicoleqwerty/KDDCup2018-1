```python
import seaborn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
air_quality.isnull().any()
```

```python
air_quality = air_quality.dropna(thresh=3)
# air_quality.dropna(how = 'all')
air_quality
```

```python
na_area = air_quality[air_quality.isnull().values == True]
for i in na_area.columns[2:]:
    print(i, len(na_area[na_area[i].isnull() == True]))
```

```python
# air_quality = air_quality.fillna(0)
# fill the NA with 0
# fill the NA with mean , 'PM10', 'NO2', 'CO', 'O3', 'SO2'
air_quality[['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'SO2']] = air_quality[['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'SO2']].fillna(air_quality[['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'SO2']].mean())
# air_quality.describe()
air_quality.head()
```
